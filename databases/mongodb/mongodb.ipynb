```python
from pymongo import MongoClient
```

```python
client = MongoClient("mongodb://127.0.0.1:27017")

db = client['test']
```

```python
db.restaurants.count()
```

```python
# 查询 相当于RDBMS中的select *
with db.restaurants.find() as cursor:
    for document in cursor:
        print type(document)
```

```python
# 查询 相当于RDBMS中的select * from restaurants where borough='Manhattan'
with db.restaurants.find({"borough": "Manhattan"}) as cursor:
    for document in cursor:
        print document['_id']
```

```python
# 查询 文档内嵌套的内容  二级查询
with db.restaurants.find({"address.zipcode": "10075"}) as cursor:
    for document in cursor:
        print document['_id']
```

```python
with db.restaurants.find({"grades.grade": "B"}) as cursor:
    for document in cursor:
        print document['grades']
```

```python
with db.restaurants.find({"grades.score": {"$gt": 30}}) as cursor:
    for document in cursor:
        print document['date']
```

```python
with db.restaurants.find({"cuisine": "Italian", "address.zipcode": "10075"}) as cursor:
    for document in cursor:
        print document
```

```python
with db.restaurants.find({"$or": [{"cuisine": "Italian"}, {"address.zipcode": "10075"}]}) as cursor:
    for document in cursor:
        print document
```

```python
import pymongo

cursor = db.restaurants.find().sort([
    ("borough", pymongo.ASCENDING),
    ("address.zipcode", pymongo.ASCENDING)
])

for document in cursor:
    print document
```

```python
result = db.restaurants.update_one(
    {"name": "Juni"},
    {
        "$set": {
            "cuisine": "American (New)"
        },
        "$currentDate": {"lastModified": True}
    }
)
```

```python
result = db.restaurants.delete_many({"borough": "Manhattan"})
```

```python

# 相当于RDBMS中的 select borough, count(1) from restaurants group by borough
cursor = db.restaurants.aggregate(
    [
        {"$group": {"_id": "$borough", "count": {"$sum": 1}}}
    ]
)
for document in cursor:
    print(document)
```

```python
# 相当于RDBMS中的select address.zipcode, count(1) from restaurants where borough='Queens' and cuisine='Brazilian'
cursor = db.restaurants.aggregate(
    [
        {"$match": {"borough": "Queens", "cuisine": "Brazilian"}},
        {"$group": {"_id": "$address.zipcode", "count": {"$sum": 1}}}
    ]
)
for document in cursor:
    print(document)
```
